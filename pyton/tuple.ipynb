8. Operations performed on tuple:
	1. all(): Returns true if all element are true or if tuple is empty. 
	2. any(): Return true if any element of the tuple is true. if tuple is empty, return false. 
	3. len(): Returns length of the tuple or size of the tuple.
	4. enumerate(): Returns enumerate object of tuple.
	5. max(): Return maximum element of given tuple. 
	6. min(): Return minimum element of given tuple. 
	7. sum(): Sums up the numbers in the tuple. 
	8. sorted(): input elements in the tuple and return a new sorted list.

```python
tuple1=(2,3,4,5,6,9,2,4,"",'null',0,7.9)
tuple1
```
