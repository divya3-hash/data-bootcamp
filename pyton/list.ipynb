7. Operations performed on list:
	1. Append(): Add an element to the end of the list. Eg: List1.append('orange')
	2. Extend(): Add all elements of a list to another list. Eg: List1.extend(['grape', 'melon'])
	3. Insert(): Insert an item at the defined index. Eg: List1.insert(1,'kiwi')
	4. Remove(): Removes an item from the list. Eg: List.remove('banana')
	5. Pop(): Removes and returns an element at the given index. Eg: List1.pop(2)
	6. Clear(): Removes all items from the list.
	7. Index(): Returns the index of the first matched item. 
	8. Count(): Returns the count of number of items passed as an argument.
	9. Sort(): Sort items in a list in ascending order 
	10. Reverse(): Reverse the order of items in the list 
	11. copy(): Returns a copy of the list
